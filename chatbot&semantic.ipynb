```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from sentence_transformers import SentenceTransformer, util
import torch

# ========== 1. microsoft/DialoGPT-medium (Chatbot Generation) ==========
print("\n==== Chatbot: DialoGPT-medium ====")

chatbot_model = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(chatbot_model)
model = AutoModelForCausalLM.from_pretrained(chatbot_model)

# Example input for chatbot
user_input = "Hello, how are you?"
input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors="pt")

# Generate a response
chat_output_ids = model.generate(input_ids, max_length=50, pad_token_id=tokenizer.eos_token_id)
response = tokenizer.decode(chat_output_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

print("User:", user_input)
print("Bot:", response)


# ========== 2. sentence-transformers/all-MiniLM-L6-v2 (Semantic Similarity) ==========
print("\n==== Semantic Similarity: all-MiniLM-L6-v2 ====")

semantic_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Sample sentences
sentence1 = "A man is playing guitar."
sentence2 = "A person is performing music."

# Encode both sentences
embedding1 = semantic_model.encode(sentence1, convert_to_tensor=True)
embedding2 = semantic_model.encode(sentence2, convert_to_tensor=True)

# Compute cosine similarity
similarity = util.pytorch_cos_sim(embedding1, embedding2)
print(f"Sentence 1: {sentence1}")
print(f"Sentence 2: {sentence2}")
print(f"Cosine Similarity Score: {similarity.item():.4f}")
```
